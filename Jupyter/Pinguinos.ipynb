```python
import math
import numpy as np
import seaborn as sns
```

```python
w1 = np.array([
[-0.617480316860551,0.836594090247448],
[-0.241471376629997,0.866835423228332],
[0.954982703113106,1.66822378060024],
[-1.38003278517147,1.06165141016727],
[0.310047949874596,0.232217517395054]

])
```

```python
w2 = np.array([
[1.03581264777408,-1.17123591052688,-0.177867633208307],
[0.366658179723632,1.31776748667085,-0.272265133299322],
[0.650429366201448,1.02399524542726,-2.73409385100892]
])
```

```python
w3 = np.array([
[-1.16696662730532,-1.274492316761,0.133140367371249],
[2.17162165919674,1.46625759841173,1.37909696462016],
[-0.356711409332845,0.293247856351775,-1.31834236999308],
[-1.22660403414976,-1.09561248021604,0.228099701985725]

])
```

```python
def f_act(X):
    activada = []
    for x in X:
        try: 
            r = 1 / (1 + math.exp(-x))
        except OverflowError:
            r = 0
        activada.append(r)
    return np.array(activada)
```

```python
#Carga de datos
iris =sns.load_dataset('penguins')
iris
```

```python
xcols = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
especie = iris['species'].unique()
especie
```

```python
X = iris[xcols].copy()
X.insert(0,'bias',1)
X
```

```python
#Prediccion 
prediccion = []
for index, fila in X.iterrows():
    capa1 = f_act(fila.dot(w1))
    capa1 = np.insert(capa1,0,1)
    capa2 = f_act(capa1.dot(w2))
    capa2 = np.insert(capa2,0,1)
    salida = f_act(capa2.dot(w3))
    prediccion.append(especie[np.argmax(salida)])
```

```python
iris['Prediccion'] = prediccion
iris.head()
```

```python
errores = iris[iris['species']!=iris['Prediccion']]
errores
```

```python
eficiencia = (1 - len(errores)/len(iris))*100
eficiencia
```
